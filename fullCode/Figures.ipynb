```python
# import modules
import os 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
print("modules imported")

figWS = "D:/ASOML/Rockies/figures/final_figures/"
summaryModel = r"D:\ASOML\Rockies\modelOutputs\fromAlpine\Rockies_modelSummary_stats.csv"
```

```python
# bar and whiskers plot of RMSE, MSE, and MAE of the different architectures
sum_df = pd.read_csv(summaryModel)

metrics = ['RMSE', 'MSE', 'MAE']

# Get unique architectures and define a color palette
architectures = sum_df['Architecture'].unique()
palette = sns.color_palette("Set2", len(architectures))
arch_colors = dict(zip(architectures, palette)) 

# Set up the figure and axes
fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharey=False)

# Loop through each metric and plot
for i, metric in enumerate(metrics):
    sns.boxplot(data=sum_df, x='Architecture', y=metric, ax=axes[i], palette=arch_colors)
    axes[i].set_title(f'{metric} by Architecture')
    axes[i].set_xlabel('Architecture')
    axes[i].set_ylabel(metric)
    axes[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.savefig("ModelArchiectures_BoxWhiskers.png", dpi=300)
plt.show()
```
